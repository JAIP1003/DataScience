```python
df=pd.read_csv("weather_data.csv",parse_dates=["day"])
df
```

```python
# Now in above table, day converted into date column
type(df.day[0])
```

```python
# To make day as my index
df.set_index('day',inplace=True)
df
```

```python
# To change the cell having NaN value with some other value
new_df=df.fillna(0)
new_df
```

```python
# If we want to update different column NaN value with different value
# By using dictionary
new_df=df.fillna({
    "day":0,
    "temperatur":0,
    "windspeed":0,
    "event":'no event'
    
})
new_df
```

```python
# If we want to fill all NaN value with previous value
new_df=df.fillna(method="ffill")
new_df
```

```python
# To copy next day value instead of previous value
new_df=df.fillna(method="bfill")
new_df
```

```python
# if we want to copy value horizontal column
new_df=df.fillna(method="ffill",axis='columns')
new_df
```

```python
# if we want to copy value horizontal column
new_df=df.fillna(method="bfill",axis='columns')
new_df
```

```python
# If we want to copy previous value only once in next NaN cell
new_df=df.fillna(method="ffill",limit=1)
new_df
```

```python
# If we want fill NaN cell with better relative value....use interpolation
new_df=df.interpolate() # By default it in linear interpolation...using method we can use quadratic,cubic etc
new_df
```

```python
# Interpolate the guess with time
new_df=df.interpolate(method='time')
new_df
```

```python
# If we want to drop the row having any NaN value
new_df=df.dropna()
new_df
```

```python
# If we want to drop the row which have all NaN value
new_df=df.dropna(how='all')
new_df
```

```python
# if we want to keep all row having atleast one value
new_df=df.dropna(thresh=1)
new_df
```

```python
# if we want to keep all row having atleast one value
new_df=df.dropna(thresh=2)
new_df
```

# Lec-6:  replace function

```python
import pandas as pd
import numpy as np
df=pd.read_csv("weather_data2.csv")
df
```

```python
new_df=df.replace(-99999,np.NaN)
new_df
```

```python
# If we want to replace two or more type of value... we will list
new_df=df.replace([-99999,-88888,7777],np.NaN)
new_df
```

```python
# To replace value based on specific column....using dictionary
new_df=df.replace({
    "temperature":-99999,
    "windspeed":-99999,
    "event":'0'
},np.NaN)
new_df
```

```python
new_df=df.replace({
    -99999:np.NaN,
    "No Event":'Sunny'
})
new_df
```

```python
#If  data set have data as well as its unit of measure and we want to remove it
new_df=df.replace('[A-Za-z]','',regex=True)
new_df

```

```python
# skip the event column as it alphabate
new_df=df.replace({
    'temperature':'[A-Za-z]',
    'windspeed':'[A-Za-z]',
    
},'',regex=True)
new_df
```
