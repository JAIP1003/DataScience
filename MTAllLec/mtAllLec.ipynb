# Lec-3: Axes labels, legend , grid

```python
day=[1,2,3,4,5,6,7]
max_t=[51,54,45,48,49,52,56]
min_t=[34,36,38,42,42,37,38]
avg_t=[40,42,42,44,45,47,51]
plt.plot(day,max_t)
```

```python
# If we want to plot all three in same graph
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Title')
plt.plot(day,max_t,label="max")
plt.plot(day,min_t,label="min")
plt.plot(day,avg_t,label="avg")

plt.legend()
```

```python
# To change the location of legend
# If we want to plot all three in same graph
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Title')
plt.plot(day,max_t,label="max")
plt.plot(day,min_t,label="min")
plt.plot(day,avg_t,label="avg")

plt.legend(loc="upper right",shadow=True, fontsize="large")

# By default , loc="best"
```

```python
#  to put grid
# If we want to plot all three in same graph
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Title')
plt.plot(day,max_t,label="max")
plt.plot(day,min_t,label="min")
plt.plot(day,avg_t,label="avg")

plt.legend()
plt.grid()
```

# Lec-4: Bar Chart

```python
import numpy as np
company=['Google','Amazon','Samsung','Microsoft','Axtria']
revenue=[80,78,56,65,30]
plt.xlabel('Company')
plt.ylabel('revenue')
plt.title("Company vs Revenue")
plt.plot(company,revenue)
```

```python
plt.xlabel('Company')
plt.ylabel('Revenue')
plt.title("USA Tech Stocks")
plt.bar(company,revenue,label="Revenue")
plt.legend()
```

```python
# If we want to plot two or more relation side by side
profit=[40,35,20,25,15]
plt.xlabel('Company')
plt.ylabel('Revenue')
plt.title("USA Tech Stocks")
plt.bar(company,revenue,label="Revenue")
plt.bar(company,profit,label="Profit")
plt.legend()
```

```python
profit=[40,35,20,25,15]
plt.xlabel('Company')
plt.ylabel('Revenue')
plt.title("USA Tech Stocks")
plt.bar(xpos,revenue,label="Revenue")
plt.bar(xpos,profit,label="Profit")
plt.legend()
```

```python
# Above plot is ploting on the same but we want to print side by side
xpos=np.arange(len(company))
plt.xticks(xpos,company)
profit=[40,35,20,25,15]
plt.xlabel('Company')
plt.ylabel('Revenue')
plt.title("USA Tech Stocks")
plt.bar(xpos,revenue,label="Revenue")
plt.bar(xpos,profit,label="Profit")
plt.legend()
```

```python
xpos=np.arange(len(company))
plt.xticks(xpos,company)
profit=[40,35,20,25,15]
plt.xlabel('Company')
plt.ylabel('Revenue')
plt.title("USA Tech Stocks")
plt.bar(xpos-0.2,revenue,label="Revenue")
plt.bar(xpos+0.2,profit,label="Profit")
plt.legend()
```

```python
xpos=np.arange(len(company))
plt.xticks(xpos,company)
profit=[40,35,20,25,15]
plt.xlabel('Company')
plt.ylabel('Revenue')
plt.title("USA Tech Stocks")
plt.bar(xpos-0.2,revenue,label="Revenue",width=0.4)
plt.bar(xpos+0.2,profit,label="Profit",width=0.4)
plt.legend()
```

```python
# If we want to show horizontal bar chart
xpos=np.arange(len(company))
plt.yticks(xpos,company)
profit=[40,35,20,25,15]
plt.xlabel('Company')
plt.ylabel('Revenue')
plt.title("USA Tech Stocks")
plt.barh(xpos-0.2,revenue,label="Revenue")
plt.barh(xpos+0.2,profit,label="Profit")
plt.legend()
```

# Lec-5: Histograms

```python
blood_suger=[115,85,90,110,100,88,93,115,125,88,77,82,129]
plt.hist(blood_suger)

#b by default it 10bin
```

```python
# To change the range
plt.hist(blood_suger,bins=3,)
```

```python
plt.hist(blood_suger, bins=3,rwidth=0.95)
```

```python
# If we want to plot two type of histo side by side
blood_suger_men=[115,85,90,110,100,88,93,115,125,88,77,82,129]
blood_suger_women=[125,75,95,120,110,58,83,105,125,88,77,84,119]
plt.hist(blood_suger_men,bins=3)
plt.hist(blood_suger_women, bins=3)
```

```python
plt.xlabel("Suger Range")
plt.ylabel("Frequecy")
plt.title("Blood Suger Analysis")
plt.hist([blood_suger_men,blood_suger_women],bins=3,color=['g','r'],label=['men','women'])
plt.legend()
```

```python
# To change the orientation of histogram
plt.xlabel("Suger Range")
plt.ylabel("Frequecy")
plt.title("Blood Suger Analysis")
plt.hist([blood_suger_men,blood_suger_women],bins=3,color=['g','r'],label=['men','women'],orientation='horizontal')
plt.legend()
```

# Lec-6: Pie chart

```python
exp_val=[1400,600,300,410,250]
exp_labels=["Home rent","Food","Phone/Internate bill","Car","Other Utilities"]
plt.axis("equal")
plt.pie(exp_val,labels=exp_labels)
```

```python
plt.pie(exp_val,labels=exp_labels)
plt.show() # To get rid of extra commnet on the top
```

```python
# If we want to increase the size of pie chart
plt.pie(exp_val,labels=exp_labels,radius=2)
plt.show() 
```

```python
# If we want  to show percentage

plt.pie(exp_val,labels=exp_labels,radius=2 ,autopct='%f%%')
plt.show() 
```

```python
# If we want to limit the decimal place
plt.pie(exp_val,labels=exp_labels,radius=2 ,autopct='%0.2f%%')
plt.show() 
```

```python
plt.pie(exp_val,labels=exp_labels,radius=2 ,autopct='%f%%',shadow=True)
plt.show() 
```

```python
plt.pie(exp_val,labels=exp_labels,radius=1.5 ,autopct='%f%%',shadow=True,explode=[0,.2,0,0.3,0])
plt.show() 
```

```python
# If we want to change the start angle
plt.pie(exp_val,labels=exp_labels,radius=1.5 ,autopct='%f%%',shadow=True,explode=[0,.2,0,0.3,0],startangle=45)
plt.show() 
```
